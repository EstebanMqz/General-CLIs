```python
import os
import requests
from bs4 import BeautifulSoup
import zipfile
import io
```

```python
url = 'https://cran.r-project.org/bin/windows/contrib/4.5/'
download_dir = '/tmp/cran_windows_packages_4_5'
os.makedirs(download_dir, exist_ok=True)

response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')
```

```python
zip_files = [link.get('href') for link in soup.find_all('a') if link.get('href').endswith('.zip')]
print(f'Found {len(zip_files)} zip files.')

for zip_file in zip_files:
    file_url = url + zip_file
    print(f'Downloading {file_url}')
    r = requests.get(file_url)
    with zipfile.ZipFile(io.BytesIO(r.content)) as z:
        extract_path = os.path.join(download_dir, zip_file.replace('.zip', ''))
        os.makedirs(extract_path, exist_ok=True)
        z.extractall(extract_path)
    print(f'Extracted {zip_file} to {extract_path}')

print('All files downloaded and extracted.')
```
